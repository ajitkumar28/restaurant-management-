```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Restaurant data
data = {
    "Item": ["Pizza", "Burger", "Pasta", "Coffee"],
    "Orders": [120, 90, 60, 150],
    "Price": [250, 120, 180, 80]
}

df = pd.DataFrame(data)

# Total sales calculation
df["Total_Sales"] = df["Orders"] * df["Price"]
print(df)

# Total revenue
revenue = np.sum(df["Total_Sales"])
print("Total Revenue:", revenue)

# Visualization
plt.bar(df["Item"], df["Total_Sales"],color='pink')
plt.xlabel("Food Item")
plt.ylabel("Total Sales")
plt.title("Restaurant Sales Performance")
plt.show()
```
